```python
import time

AUTH_TOKEN = "fe5171e1db6b6eeaa8e5a68e200f51dd5ec6e1db"

SINCE_DATE = "22024-09-22"
UNTIL_DATE = "2024-12-16"
LANGUAGE = "id"

TOTAL_LIMIT = 2000
BATCH_SIZE = 100
SLEEP_TIME = 120

collected = 0
batch = 1
KEYWORD = "prabowo"

while collected < TOTAL_LIMIT:
    filename = f"tweets_prabowo_part{batch}.csv"
    query = f"{KEYWORD} lang:{LANGUAGE} since:{SINCE_DATE} until:{UNTIL_DATE}"

    print(f"Gibran batch {batch}")

    !npx tweet-harvest@2.6.1 -o "{filename}" -s "{query}" -l {BATCH_SIZE} --token "{AUTH_TOKEN}"

    collected += BATCH_SIZE
    batch += 1

    if collected < TOTAL_LIMIT:
        time.sleep(SLEEP_TIME)
```

```python
import pandas as pd
import glob

files = glob.glob("/content/tweets-data/tweets_*.csv")

df_list = []

for f in files:
    df = pd.read_csv(f)
    key = f.split("_")[1]
    df["keyword"] = key
    df_list.append(df)

final_df = pd.concat(df_list, ignore_index=True)

final_df.to_csv("dataset_politik_10000.csv", index=False, encoding="utf-8-sig")

print("Total data:", len(final_df))
print(final_df["keyword"].value_counts())
```
