```python
import pandas as pd
import glob

# Mencari semua file .csv yang diawali dengan 'tweets_'
files = glob.glob("/content/tweets-data/tweets_*.csv")
print(f"File yang akan digabungkan: {files}")

df_list = []
for f in files:
    try:
        # Menambahkan kolom 'aspek' berdasarkan nama file
        aspect_name = f.replace('tweets_', '').replace('.csv', '')
        df_file = pd.read_csv(f)
        df_file['aspek'] = aspect_name # Menandai data ini dari aspek mana
        df_list.append(df_file)
        print(f"File {f} dimuat, {len(df_file)} baris, diberi label aspek: {aspect_name}")

    except pd.errors.EmptyDataError:
        print(f"Peringatan: File {f} kosong dan akan dilewati.")
    except Exception as e:
        print(f"Gagal memproses file {f}: {e}")

if df_list:
    # Menggabungkan semua dataframe
    df_merged = pd.concat(df_list, ignore_index=True)

    # Menyimpan ke file CSV gabungan
    output_filename = "merged_tweets_2024.csv"
    df_merged.to_csv(output_filename, index=False, encoding="utf-8-sig")

    print(f"\nBerhasil! Semua data telah digabung ke: {output_filename}")
    print(f"Total baris data terkumpul: {len(df_merged)}")

    # Menampilkan distribusi data per aspek
    print("\nDistribusi data per aspek:")
    print(df_merged['aspek'].value_counts())
else:
    print("Tidak ada data untuk digabungkan.")
```
